```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect('5784/noahs.sqlite')
conn.row_factory = sqlite3.Row

cur = conn.cursor()

res = cur.execute('SELECT * FROM customers')
customers_data = [dict(row) for row in res.fetchall()]

cur.close()
conn.close()
```

```python
customers_df = pd.DataFrame(columns=customers_data[0].keys(), 
                            data=customers_data)
```
